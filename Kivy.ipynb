```python
#MULTIPLE PLOTS
#subplots() draw multiple plots in 1 figure
# .subplot(1,2,1) the figure has 1 row, 2 columns, and this plot is the first plot.
# .subplot(1,2,2) the figure has 1 row, 2 columns, and this plot is the 2nd plot.
# title   title()
# super title  suptitle()
```

```python
import matplotlib.pyplot as plt
import numpy as nm
#plot 1
x=np.array([5,54,33])
y=np.array([34,3,23])
plt.subplot(1,2,1)
plt.plot(x,y)
# .subplot(1,2,1) the figure has 1 row, 2 columns, and this plot is the first plot.
#plot 2
x=np.array([4,23,8,9])
y=np.array([2,43,2,1])
plt.subplot(1,2,2)
plt.plot(x,y)
plt.show()
```

```python
# with 2 rows an 1 column

import matplotlib.pyplot as plt
import numpy as np
#plot 1
x=np.array([5,54,33])
y=np.array([34,3,23])

plt.subplot(2,2,1) # 2 row 2 col 1st plot
plt.plot(x,y)
plt.title("SALES")

#plot 2
x=np.array([4,23,8,9])
y=np.array([2,43,2,1])

plt.subplot(2,2,2) # 2 row 2 col 2nd plot
plt.plot(x,y)
plt.title("DATA")

#plot 3
x=np.array([4,20,8,9])
y=np.array([2,3,2,1])

plt.subplot(2,2,3) #2 row 2 col 3rd plot
plt.plot(x,y)
plt.title("INCOME")

# plot 4
x=np.array([4,2,4,9])
y=np.array([2,31,2,1])

plt.subplot(2,2,4) #2 row 2 col 4th plot
plt.scatter(x, y)
plt.title("NETWORKING")

plt.suptitle("DATA ANALISIS")
plt.show()
```

```python
# LABELS 
import matplotlib.pyplot as plt
import numpy as np
y=np.array([34,12,30,20])
l=['apples','banana','cherries','dates']
plt.pie(y, labels=l)
plt.suptitle("Fruits")
plt.show()
```

```python
#default start from x axis
# we can change by using  'startangle' parameter 

import matplotlib.pyplot as plt
import numpy as np
y=np.array([34,12,30,20])
l=['apples','banana','cherries','dates']
plt.pie(y, labels=l,startangle=180)
plt.suptitle("Fruits")
plt.show()
```

```python
# explode
# syntax   myexplode=[0.2, 0.2, 0, 0]
# 0.2 from the center of the pie

import matplotlib.pyplot as plt
import numpy as np
y=np.array([34,12,30,20])
l=['apples','banana','cherries','dates']
myexplode=[0.1, 0, 0.2, 0]
plt.pie(y, labels=l,startangle=90, explode=myexplode)

plt.suptitle("Fruits")
plt.show()
```

```python
# shadow parater to True
# legend() -- to add a list of explanation for each wedge
# plt.legend(title='Four fruits')

import matplotlib.pyplot as plt
import numpy as np
y=np.array([34,12,30,20])
l=['apples','banana','cherries','dates']
myexplode=[0.2,0,0,0]
plt.pie(y, labels=l,startangle=90, shadow=True, explode=myexplode, autopct='%1.1f%%')
plt.legend()
plt.suptitle("Fruits")
plt.show()
```
